```python
# Proceso de limpieza del archivo Nivel1
datos1_df = pd.read_excel('C:/Users/Madai/Documents/GitHub/PC3/appendData/level1.xlsx', header=4)
# Eliminar las filas 1 y 2
datos1_limpios = datos1_df.drop([0, 1])
# Eliminar columnas que no tienen nombre (sin encabezados)
datos1_limpios = datos1_limpios.loc[:, ~datos1_limpios.columns.str.contains('^Unnamed')]
datos1_limpios.reset_index(drop=True, inplace=True)
print("Nivel 1 - Datos Limpios:\n", datos1_limpios.head())
```

```python
# Proceso de limpieza del archivo Nivel2
datos2_df = pd.read_excel('C:/Users/Madai/Documents/GitHub/PC3/appendData/level2.xlsx', header=0)
# Eliminar las filas 2 y 3
datos2_limpios = datos2_df.drop([0, 1])
# Eliminar columnas que no tienen nombre (sin encabezados)
datos2_limpios = datos2_limpios.loc[:, ~datos2_limpios.columns.str.contains('^Unnamed')]
datos2_limpios.reset_index(drop=True, inplace=True)
print("Nivel 2 - Datos Limpios:\n", datos1_limpios.head())
```

```python
# Proceso de limpieza del archivo Nivel3
datos3_df = pd.read_excel('C:/Users/Madai/Documents/GitHub/PC3/appendData/level3.xlsx', header=0)
# Eliminar las filas 2 y 3
datos3_limpios = datos3_df.drop([0, 1])
datos3_limpios = datos3_limpios.loc[:, ~datos3_limpios.columns.str.contains('^Unnamed')]
# Reiniciar los índices después de eliminar las filas
datos3_limpios.reset_index(drop=True, inplace=True)
print("Nivel 3 - Datos Limpios:\n", datos3_limpios.head())
```

```python
# Proceso de limpieza del archivo Nivel3
datos4_df = pd.read_excel('C:/Users/Madai/Documents/GitHub/PC3/appendData/level4.xlsx', header=0)
#Se elimina las columnas completamente vacias
datos4_df.dropna(how='all', axis=1, inplace=True)
datos4_limpios = datos4_df.loc[:, ~datos4_df.columns.str.contains('^Unnamed')]
print("Nivel 4 - Datos Limpios:\n", datos4_limpios.head())
```

```python
# Combinación de todos los DataFrames en uno solo
datos_combinados = pd.concat([datos1_limpios, datos2_limpios, datos3_limpios, datos4_limpios], ignore_index=True)
print(datos_combinados.head())
```

```python
import pandas as pd

# Cargamos el archivo CSV que contiene los datos de fallecidos
ruta_fallecidos = 'C:/Users/Madai/Documents/GitHub/PC3/fallecidosCOVID.csv'
datos_covid = pd.read_csv(ruta_fallecidos, sep=';')

# Convertimos la columna 'FECHA_FALLECIMIENTO' a un formato de fecha
datos_covid['FECHA_FALLECIMIENTO'] = pd.to_datetime(datos_covid['FECHA_FALLECIMIENTO'], format='%Y%m%d', errors='coerce')

# Agrupamos los datos por mes y año, contando el número de fallecidos en cada periodo
fallecidos_mensuales = datos_covid.groupby(datos_covid['FECHA_FALLECIMIENTO'].dt.to_period('M')).size().reset_index(name='Total_Fallecidos')

# Mostramos el reporte mensual de fallecidos en el formato deseado
for i, fila in fallecidos_mensuales.iterrows():
    print(f"Mes y Año: {fila['FECHA_FALLECIMIENTO']}, Total Fallecidos: {fila['Total_Fallecidos']}")
```

```python
import pandas as pd

# Cargamos el archivo CSV de datos de fallecidos
ruta_fallecidos = 'C:/Users/Madai/Documents/GitHub/PC3/fallecidosCOVID.csv'
datos_covid = pd.read_csv(ruta_fallecidos, sep=';')

# Convertimos la columna de fecha de fallecimiento a formato datetime
datos_covid['FECHA_FALLECIMIENTO'] = pd.to_datetime(datos_covid['FECHA_FALLECIMIENTO'], format='%Y%m%d', errors='coerce')

# Eliminamos espacios en blanco al inicio y al final en la clasificación de muerte
datos_covid['CLASIFICACION_DEF'] = datos_covid['CLASIFICACION_DEF'].str.strip()

# Agrupamos los datos por mes y tipo de muerte, contando el número de fallecidos
reporte_mensual = datos_covid.groupby([datos_covid['FECHA_FALLECIMIENTO'].dt.to_period('M'), 'CLASIFICACION_DEF']).size().unstack(fill_value=0)

# Mostramos el informe de fallecidos por mes y clasificación de muerte
print(reporte_mensual)
```
